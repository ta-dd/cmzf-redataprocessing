```python
import pkgutil

package = rdp
prefix = package.__name__ + "."
for importer, modname, ispkg in pkgutil.iter_modules(package.__path__, prefix):
    print("Found submodule %s (is a package: %s)" % (modname, ispkg))
    module = __import__(modname, fromlist="dummy")
    print("Imported", module)
```

**Load the data from SQLite:**

```python
path_to_sqlite='estate_data.sqlite'
con = sqlite3.connect(path_to_sqlite)
```

```python
print("Tables in our database:\n")
cursor = con.cursor()
# Getting all tables from sqlite_master
sql_query = """SELECT name FROM sqlite_master WHERE type='table';"""
cursor.execute(sql_query)
sqlite_tables = cursor.fetchall()
sqlite_tables = [*map(lambda table: table[0] ,sqlite_tables)]
print(sqlite_tables)
del sqlite_tables
del sql_query
```

```python
import pandas as pd
```

```python
df_description = pd.read_sql('SELECT * FROM DESCRIPTION_APARTMENTS_SALE', con = con)
```
