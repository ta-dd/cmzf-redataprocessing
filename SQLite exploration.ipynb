```python
import pandas as pd
import sqlite3
```

```python
path = "Prague_apartments_sale_2023-01-14.gzip"
df = pd.read_parquet(path)
df.head()
```

```python
path_2 = "Prague_apartments_sale_2023-01-14_desc.gzip"
df_2 = pd.read_parquet(path_2)
df_2.head()
```

##### **SQLite:**

```python
con = sqlite3.connect('DB.sqlite')
```

```python
df.to_sql(name = 'df', con= con, index = False)
```

```python
df_2.to_sql(name = 'df_2', con= con, index = False)
```

```python
del df
del df_2
```

```python
# load again from the storage
df = pd.read_sql('SELECT * FROM df', con = con)
df_2 = pd.read_sql('SELECT * FROM df_2', con = con)
```

```python
# Creating a cursor object using the cursor() method
cursor = con.cursor()

# Getting all tables from sqlite_master
sql_query = """SELECT name FROM sqlite_master WHERE type='table';"""
cursor.execute(sql_query)

print(cursor.fetchall())
#cursor.fetchall()[0][0]

# Delete table if already exists
#cursor.execute("DROP TABLE df")
#cursor.execute("DROP TABLE df_2")

# Commit your changes in the database
con.commit()
```

```python
df_copy = df.copy()
```

```python
# Appending to the table - columns must be the same
df_copy.to_sql(name = 'df', con= con, index = False, if_exists='append')
```

```python
# Closing the connection
con.close()
```
